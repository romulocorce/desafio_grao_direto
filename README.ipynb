- **Desenvolvido por:** Rômulo Cristovam Corce
- **Tecnologias:** Databricks, PySpark, Delta Lake, Unity Catalog
- **Dataset:** grain_logistic_shipping.csv
### **Objetivo**

Estruturar um pipeline de ingestão, transformação e análise de dados logísticos com base no conceito de arquitetura Medallion (Bronze, Silver, Gold) utilizando a plataforma Databricks. O foco é garantir a governança dos dados, qualidade e disponibilidade para consumo analítico.

## **Arquitetura Implementada**

### **Camada Bronze - Ingestão**
- Leitura do arquivo CSV original, com preservação total do formato.
- Registro de metadados da ingestão: data/hora e path do arquivo.
- Armazenamento em formato Delta no Unity Catalog.

### **Camada Silver - Limpeza e Padronização**
- Conversão de tipos de dados conforme o dicionário de dados.
- Normalização dos estados brasileiros com fallback para "Estado inexistente".
- Conversão de datas de envio e entrega para o formato ISO.
- Tratamento de valores ausentes com preenchimento padrão.
- Padronização de categorias: "importância", "gênero", "chegou_no_tempo".
- Cálculo da coluna "tempo_entrega_dias".
- Remoção de duplicatas utilizando a chave "id_envio".
- Adição de metadados: camada e data de processamento.
- Validação de Dados

### **Testes Automatizados**
Durante a transformação Silver, foram aplicadas validações automáticas de:
- Conformidade de schema;
- Existência de valores nulos em campos obrigatórios;
- Verificação de domínio em colunas categóricas.
- Consumo e Visualização

### **Camada Gold - Agregações e Métricas**
Foram criadas múltiplas tabelas analíticas com foco em diferentes perspectivas:

- **gold_desempenho_por_metodo_envio.**
> -  `O que demonstra:` Performance dos métodos de envio com base em volume de envios, satisfação do cliente, tempo médio de entrega e percentual de atrasos.
> -  `Qual pergunta responde:` Quais métodos de envio performam melhor em termos de satisfação e pontualidade?
> -  `Motivo de uso:` Ajuda na avaliação de fornecedores logísticos e priorização de canais mais eficientes.

- **gold_satisfacao_por_destino.**
> -  `O que demonstra:` Avaliação média dos clientes e da entrega por localidade de destino.
> -  `Qual pergunta responde:` Para quais destinos os clientes estão mais satisfeitos com o serviço?
> -  `Motivo de uso:` Identifica gargalos e oportunidades de melhoria regional na operação de entrega.

- **gold_analise_mensal_envios.**
> -  `O que demonstra:` Evolução mensal de volume de envios, receita gerada, peso transportado e descontos aplicados.
> -  `Qual pergunta responde:` Como os indicadores logísticos e financeiros variaram mês a mês?
> -  `Motivo de uso:` Permite acompanhar sazonalidade, crescimento e impacto de campanhas promocionais.

-** gold_relacao_peso_custo.**
> -  `O que demonstra:` Relação entre o peso dos envios e o custo, com correlação estatística.
> -  `Qual pergunta responde:` O peso impacta diretamente o custo de envio?
> -  `Motivo de uso:` Suporte à precificação de frete e análises de eficiência operacional por tipo de carga.

- **gold_produtividade_mensal_comparativo.**
> -  `O que demonstra:` Variação mensal da produtividade e receita, com cálculo de crescimento/queda.
> -  `Qual pergunta responde:` Estamos crescendo em volume e receita? Qual a tendência comparada ao mês anterior?
> -  `Motivo de uso:` Facilita decisões de gestão e planejamento com base em tendências operacionais.

- **gold_satisfacao_vs_desempenho.**
> -  `O que demonstra:` Correlação entre a pontualidade da entrega, a importância do cliente e métricas de satisfação.
> -  `Qual pergunta responde:` Clientes importantes que receberam com atraso estão mais insatisfeitos?
> -  `Motivo de uso:` Alinha performance logística com impacto na experiência do cliente, priorizando contas estratégicas.

- **gold_eficiencia_logistica.**
> -  `O que demonstra:` Eficiência logística por método de envio e destino, com foco em tempo de entrega, variação e taxa de pontualidade.
> -  `Qual pergunta responde:` Quais rotas e métodos são mais eficientes em termos de tempo e regularidade?
> -  `Motivo de uso:` Suporta otimização de rotas, negociação com transportadoras e planejamento logístico.

- **gold_rentabilidade_por_importancia.**
> -  `O que demonstra:` Receita, desconto aplicado e ticket médio por nível de importância do cliente.
> -  `Qual pergunta responde:` Clientes mais importantes geram maior receita e têm maior rentabilidade?
> -  `Motivo de uso:` Permite avaliar o retorno por tipo de cliente, ajustar políticas comerciais e descontos.

- **gold_desempenho_por_corredor.**
> -  `O que demonstra:` Desempenho logístico por corredor de armazenagem, incluindo atrasos, satisfação e peso total.
> -  `Qual pergunta responde:` Existem áreas do estoque mais críticas para o desempenho logístico?
> -  `Motivo de uso:`  Direciona melhorias internas na armazenagem e separação de pedidos, reduzindo falhas operacionais.

As tabelas da camada Gold estão estruturadas para consumo direto por ferramentas analíticas como Power BI, notebooks interativos e dashboards dentro do próprio ambiente Databricks.
